### 1. ROC Curve and the 45-Degree Line (No-Skill Classifier)

The diagonal 45-degree line on an ROC curve represents a baseline model that performs no better than random guessing. The ROC curve is a plot of the **True Positive Rate (TPR)** against the **False Positive Rate (FPR)**:

- **True Positive Rate (TPR)**: The proportion of actual positives correctly identified:  
  $
  TPR = \frac{TP}{TP + FN}
  $

- **False Positive Rate (FPR)**: The proportion of actual negatives incorrectly identified as positive:  
  $
  FPR = \frac{FP}{FP + TN}
  $

For a no-skill classifier, the chance of predicting a positive outcome is the same as predicting a negative outcome. This means the number of true positives ($TP$) is proportional to the number of false positives ($FP$).

```python
import matplotlib.pyplot as plt
import numpy as np

# Parameters
phi = 0.3  # Proportion of positives

# ROC Curves
def plot_roc():
    plt.figure(figsize=(10, 5))
    
    # Zero Skill Classifier
    x = np.linspace(0, 1, 100)
    plt.plot(x, x, label="Zero Skill Classifier (45-degree)", linestyle='--')
    
    # Perfect Classifier
    perfect_x = [0, 0, 1]
    perfect_y = [0, 1, 1]
    plt.plot(perfect_x, perfect_y, label="Perfect Classifier")
    
    plt.title("ROC Curve")
    plt.xlabel("False Positive Rate (FPR)")
    plt.ylabel("True Positive Rate (TPR)")
    plt.legend()
    plt.grid(True)
    plt.show()

# Precision-Recall Curves
def plot_precision_recall():
    plt.figure(figsize=(10, 5))
    
    # Zero Skill Classifier
    plt.hlines(y=phi, xmin=0, xmax=1, label=f"Zero Skill Classifier (Precision={phi})", colors='orange', linestyles='--')
    
    # Perfect Classifier
    perfect_pr_x = [0, 1, 1]
    perfect_pr_y = [1, 1, phi]
    plt.plot(perfect_pr_x, perfect_pr_y, label="Perfect Classifier")
    
    plt.title("Precision-Recall Curve")
    plt.xlabel("Recall")
    plt.ylabel("Precision")
    plt.legend()
    plt.grid(True)
    plt.show()

# Plotting
plot_roc()
plot_precision_recall()
```
